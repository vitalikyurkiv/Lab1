```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = ['times_pregnant', 'plasma_glucose', 'blood_pressure', 'triceps_skin', 'serum_insulin', 'body_mass_index', 'diabetes_pedigree_function', 'age', 'class_variable']
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
dataset = pd.read_csv(url, names=column_names)
```

```python
dataset.head()
```

```python
#Визначаємо розмір датасету
print("Size of dataset:", dataset.shape)
# Перевірка на пусті значення
missing_data = dataset.isna().sum()
print(missing_data, "\n")

columns_to_replace_zero = ['plasma_glucose', 'blood_pressure', 'triceps_skin', 'serum_insulin', 'body_mass_index', 'diabetes_pedigree_function', 'age']

# Перевірка на значення , які рівню 0
zero_data = (dataset[columns_to_replace_zero] == 0).sum()
print(zero_data)

# Заміна нулів на pd.NA для відповідних стовпців
dataset[columns_to_replace_zero] = dataset[columns_to_replace_zero].replace(0, pd.NA)
```

```python
mean_values = dataset.mean()
var_values = dataset.var()
std_values = dataset.std()

print("Mean values:")
print(mean_values, "\n")
print("Variance:")
print(var_values, "\n")
print("Standard Deviation:")
print(std_values, "\n")
```

```python
# Побудова гістограм
for column in column_names:
    plt.figure(figsize=(8, 6))
    plt.hist(dataset[column].dropna(),  color='maroon')
    plt.title(f'Histogram of {column}')
    plt.grid(True)
    plt.show()
```
